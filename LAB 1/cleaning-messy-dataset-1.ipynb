```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Jaivansh Chawla\Documents\COLLEGE\ML LAB\my_file (1).csv')
```

```python
# Remove leading/trailing spaces in column names
df.columns = df.columns.str.strip() 
df.columns
```

```python
df.info()
```

```python
# Remove square bracket content (e.g. [note]) from 'Peak' column
df['Peak'] = df['Peak'].str.replace(r'\[.*?\]', '', regex=True)

# Fill missing values with empty string
df['Peak'].fillna('', inplace = True)
df
```

```python
# Remove square bracket content from 'All Time Peak' column
df['All Time Peak'] = df['All Time Peak'].str.replace(r'\[.*?\]', '', regex=True)

# Fill missing values with empty string
df['All Time Peak'].fillna('', inplace = True)
df
```

```python
# Replace multiple whitespace characters (including non-breaking spaces) with a single space, and strip
df.columns = df.columns.str.replace(r'\s+', ' ', regex=True).str.strip()
cleaned_columns = df.columns.tolist()
df.columns
```

```python
# Remove dollar signs, commas, and bracketed notes; strip whitespace
df['Actual gross'] = (
    df['Actual gross']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .str.replace(r'\[.*?\]', '', regex=True)
    .str.strip()
)
df
```

```python
#  Remove dollar signs and commas; strip whitespace
df['Adjusted gross (in 2022 dollars)'] = (
    df['Adjusted gross (in 2022 dollars)']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)  # Remove $ and commas
    .str.strip()
)
df
```

```python
# Remove footnote symbols (e.g. *, †, ‡) and bracketed notes
df['Tour title'] = df['Tour title'].str.replace(r'[\.*?‡†]', '', regex=True).str.replace(r'\[.*?\]', '', regex=True)
df
```

```python
#Split the year to start year and end year
df[['Start Year', 'End Year']] = df['Year(s)'].str.split('–', n=1, expand= True)
df
```

```python
#Remove any dollar sign, comma or any other signs
df['Average gross'] = (
    df['Average gross']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)  # Remove $ and commas
    .str.strip()
)
df
```

```python
#Save the cleaned dataset
df.to_csv("cleaned_data.csv", index=False)
```
