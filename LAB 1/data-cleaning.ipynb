```python
df = pd.read_csv(r'C:\Users\Jaivansh Chawla\Documents\COLLEGE\ML LAB\my_file (1).csv')
print("UNCLEANED DATA")
df
```

```python
df["Ref."]= df["Ref."].apply(lambda x : re.sub(r"[^a-zA-Z0-9]" , "" , x))
df.head(5)
```

```python
print([{col} for col in df.columns])
```

```python
df.columns = df.columns.str.replace('\xa0', ' ', regex=False)
print([col for col in df.columns])
```

```python
def clean_text(text):
    text = re.sub(r"\[\s*.*?\s*\]", "" , text)
    text = re.sub(r"[^a-zA-Z0-9]", "" , text)
    return text 
columns = ["Actual gross" , "Adjusted gross (in 2022 dollars)" , "Average gross"]
for column in columns:
    df[column] = df[column].apply(clean_text)
df.head(5)
```

```python
def clean_peak(text):
    if not pd.isna(text):
        text =  re.sub(r"\[\s*.*?\s*\]", "" , text)
    return text
columns = ["Peak" , "All Time Peak"]
for column in columns:
    df[column] = df[column].apply(clean_peak)
df.head(5)
```

```python
def clean_title(text):
    text =  re.sub(r"[^a-zA-Z ]", "" , text)
    return text
df["Tour title"] = df["Tour title"].apply(clean_title)
```

```python
df.head(5)
```

```python
df["Ref."] = df["Ref."].replace("1516" , "16")
df["Ref."] = df["Ref."].replace("d" , "5")
```

```python
df.head(5)
```

```python
numeric_cols = ["Actual gross" , "Adjusted gross (in 2022 dollars)" 
                , "Average gross" , "Shows" , "Ref." , "Peak" , "All Time Peak"]
for cols in numeric_cols :
    df[cols] = df[cols].astype("float")
```

```python
df = df.drop(["All Time Peak" , "Peak"] , axis =1)
print("CLEANED DATA")
df
```
